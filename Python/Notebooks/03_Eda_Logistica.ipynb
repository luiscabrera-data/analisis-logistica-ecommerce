```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Producto_Clean = pd.read_csv('../Data/Clean/Producto_Clean.csv')
Pedidos_Clean = pd.read_csv('../Data/Clean/pedidos_clean.csv')
Almacenes_Clean = pd.read_csv('../Data/Clean/Almacenes_Clean.csv')
Logistica_Clean = pd.read_csv('../Data/Clean/Logistica_Clean.csv')
```

```python
# Distribucion de estado de envio
conteo = Logistica_Clean['estado_envio'].value_counts()
print(conteo)
conteo.plot(kind='bar', title='Pedidos por estado de envío')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python
# Costo de envío por almacén
costo_almacen = Logistica_Clean.groupby('id_almacen')['costo_envio'].mean().round(2)
print(costo_almacen)
Logistica_Clean.groupby('id_almacen')['costo_envio'].mean().plot(kind='bar')
plt.title('Costo promedio por almacén')
plt.show()
```

```python
# Ver costo sin outliers
sin_outliers = Logistica_Clean[Logistica_Clean['flag_costo_atipico'] == 0]
print(sin_outliers.groupby('id_almacen')['costo_envio'].mean().round(2))
```
