```python
import pandas as pd
import numpy as np
```

# LIMPIEZA TABLA PRODUCTOS

```python
# HACEMOS UNA COPIA DE LA TABLA PRODUCTO_RAW PARA PRODUCTO_CLEAN Y LO LLAMAMOS
Producto_Raw = pd.read_csv('../Data/Raw/productos.csv')
Producto_Clean = Producto_Raw.copy()
Producto_Clean.head(10)
```

```python
# Quitar espacios en columnas de texto
Producto_Clean['nombre_producto'] = Producto_Clean['nombre_producto'].str.strip()
Producto_Clean['categoria'] = Producto_Clean['categoria'].str.strip()
Producto_Clean['precio_base'] = Producto_Clean['precio_base'].str.strip()
```

```python
# Quitar símbolo $ y convertir a número
Producto_Clean['precio_base'] = Producto_Clean['precio_base'].str.replace('$','',regex=False)
Producto_Clean['precio_base'] = pd.to_numeric(Producto_Clean['precio_base'], errors='coerce')
```

```python
# Combertir un numero negativo a positivo
Producto_Clean['precio_base'] = abs(Producto_Clean['precio_base'])
```

```python
# Remplazar la palabra mal escrita por la buena
Producto_Clean['categoria'] = Producto_Clean['categoria'].replace({
    'Elec' : 'Electrónica',
    'electtronica' : 'Electrónica',
    'Eletrónica' : 'Electrónica',
    'Electronica' : 'Electrónica',
    'eletronica' : 'Electrónica',
    'ElecTrónica' : 'Electrónica',
})
```

```python
# Remplazar el " con nada
Producto_Clean['nombre_producto'] = Producto_Clean['nombre_producto'].str.replace('"','', regex=False)
```

```python
#
Producto_Clean.to_csv('../Data/Clean/Producto_Clean.csv', index=False)
```

# LIMPIEZA TABLA PEDIDOS

```python
# HACEMOS UNA COPIA DE LA TABLA PRODUCTO_RAW PARA PRODUCTO_CLEAN Y LO LLAMAMOS
Pedidos_Raw = pd.read_csv('../Data/Raw/pedidos.csv')
Pedidos_Clean = Pedidos_Raw.copy()
Pedidos_Clean.head(20)
```

```python
# Eliminar espacios en blanco al inicio y final
Pedidos_Clean['id_cliente'] = Pedidos_Clean['id_cliente'].str.strip()
Pedidos_Clean['fecha_pedido'] = Pedidos_Clean['fecha_pedido'].str.strip()
```

```python
# Transformar los tipos de datos de al columnas
def convertir_fecha(fecha):
    for formato in ['%Y-%m-%d', '%d/%m/%Y', '%m-%d-%Y']:
        try:
            return pd.to_datetime(fecha, format=formato)
        except:
            continue
    return pd.NaT

Pedidos_Clean['fecha_pedido'] = Pedidos_Clean['fecha_pedido'].apply(convertir_fecha)
```

```python
# Eliminar duplicados
Pedidos_Clean = Pedidos_Clean.drop_duplicates()
```

```python
# Descargar el nuevo csv limpio
Pedidos_Clean.to_csv("../Data/Clean/pedidos_clean.csv", index=False)
```

# LIMPIEZA TABLA LOGISTICA

```python
# HACEMOS UNA COPIA DE LA TABLA Logistica_Raw PARA Logistica_Clean Y LO LLAMAMOS
Logistica_Raw = pd.read_csv('../Data/Raw/logistica.csv')
Logistica_Clean = Logistica_Raw.copy()
Logistica_Clean.head(10)
```

```python
# Eliminar espacios vacios
Logistica_Clean['fecha_entrega_real'] = Logistica_Clean['fecha_entrega_real'].str.strip()
Logistica_Clean['estado_envio'] = Logistica_Clean['estado_envio'].str.strip()
```

```python
# Transformar los tipos de datos de al columnas
def convertir_fecha(fecha):
    for formato in ['%Y-%m-%d', '%d/%m/%Y', '%m-%d-%Y']:
        try:
            return pd.to_datetime(fecha, format=formato)
        except:
            continue
    return pd.NaT

Logistica_Clean['fecha_entrega_real'] = Logistica_Clean['fecha_entrega_real'].apply(convertir_fecha)
```

```python
# Eliminar duplicados
Logistica_Clean = Logistica_Clean.drop_duplicates()
```

```python
# Usamos flag para costo_envio si es menor a 100 lo pone como 1 y si es mayor como 0
Logistica_Clean['flag_costo_atipico'] = (Logistica_Clean['costo_envio'] > 100).astype(int)
```

```python
Logistica_Clean.to_csv('../Data/Clean/Logistica_Clean.csv', index=False)
```
